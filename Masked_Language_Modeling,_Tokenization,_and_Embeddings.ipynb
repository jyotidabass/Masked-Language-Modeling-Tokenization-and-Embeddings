```python
# Import necessary libraries
from transformers import BertTokenizer, BertModel
import torch

# Load pre-trained model tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Example sentence for masked language modeling
sentence = "Hello, how are you?"

# Tokenize the sentence
inputs = tokenizer(sentence, return_tensors="pt")

# Get the input IDs and attention mask
input_ids = inputs["input_ids"]
attention_mask = inputs["attention_mask"]

# Mask one of the tokens (for example, the third token)
masked_index = 2
input_ids[0, masked_index] = tokenizer.mask_token_id

# Forward pass, get pooled_output and last_hidden_state
outputs = model(input_ids, attention_mask=attention_mask)

# Last hidden state
last_hidden_state = outputs.last_hidden_state

# Pooled output (representation of the sequence)
pooled_output = outputs.pooler_output

# Print the embeddings (last hidden state) of the first token (CLS token)
print("Embeddings of the CLS token:", last_hidden_state[0, 0, :])

# To predict the masked token, you would typically use the last hidden state
# of the masked token and pass it through a classification layer.
# Here's a simplified example to demonstrate how to get the embeddings of the masked token:
masked_token_embeddings = last_hidden_state[0, masked_index, :]
print("Embeddings of the masked token:", masked_token_embeddings)

# For predicting the masked token, you would use a linear layer on top of these embeddings.
# Here's how you could set up a simple linear layer for demonstration:
linear_layer = torch.nn.Linear(model.config.hidden_size, model.config.vocab_size)
# Note: This is a simplified example and actual implementation might vary based on your specific requirements.

# Example usage of the linear layer (this part is more about illustrating the concept):
output = linear_layer(masked_token_embeddings)
print("Predicted token ID (simplified example):", torch.argmax(output))
```
